```python
col_names = ["Calories","Total Fat","Carbohydrates","Dietary Fiber","Sugars","Protein","Vitamin A (% Daily Value)","Vitamin C (% Daily Value)","Calcium (% Daily Value)","Iron (% Daily Value)"]
for i in col_names:
    maximum = max(df[i])
    max_index = df[i].idxmax()
    print(i+":  "+df["Item"][max_index]+" - "+str(maximum))
```

### Question 6

```python
col_names = ["Calories","Total Fat","Carbohydrates","Dietary Fiber","Sugars","Protein","Vitamin A (% Daily Value)","Vitamin C (% Daily Value)","Calcium (% Daily Value)","Iron (% Daily Value)"]
index = 1
plt.figure(figsize=(10, 50))
for i in col_names:
    plt.subplot(10,1,index)
    index=index+1
    sns.stripplot(x = "Category", y = i, data = df)
    plt.xticks(rotation=20)
```

### Question 7

```python
cates = df["Category"].unique()
index = 1
for i in cates:
    plt.figure(figsize=(10, 90))
    df_new = df.loc[df["Category"] == i].sort_values(by="Calories",ascending = False)
    plt.subplot(len(cates),1,index)
    plt.barh(df_new["Item"],df_new["Calories"])
    index=index+1
    plt.yticks(rotation=20)
    if(i == "Coffee & Tea"):
        plt.gcf().set_size_inches(10,180)
```

##### Item calorie values in each category are varied such that all contain items that have few and many calories
